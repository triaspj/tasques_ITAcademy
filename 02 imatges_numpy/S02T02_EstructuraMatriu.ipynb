### Exercici 8 Manipulacio Imatges

****Importem llibreries****

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

****Carreguem i mostrem imatrge Imatge****

```python
imgOriginal = mpimg.imread('peix.png')
plt.imshow(imgOriginal)
```

****Mascara per eliminar el canal verd i guardar imatge****

```python
imgMask=[1,0,1,1]
imgSinG=imgOriginal*imgMask
plt.imsave('PeixNoG.png',imgSinG)
```

```python
plt.imshow(imgSinG)
```

```python
imgMask=[1,1,0,1]
imgSinB=imgOriginal*imgMask
plt.imsave('PeixNoB.png',imgSinG)
```
